```python
import pandas as pd
```

```python
# Load the file - Table 6
file_path = 'Tabla_6_diagnosis_sex_age.xlsx'
xls = pd.ExcelFile(file_path)

# Load the first sheet
df = pd.read_excel(xls, sheet_name='Hoja1')
```

```python
# Create separate dataframes for Male and Female
df_male = df[['Age Group', 'Number of diagnosis Male', 'Diagnosis Percentage Male']].copy()
df_male['Sex'] = 'Male'
df_male.columns = ['Age Group', 'Number of diagnosis', 'Ratio', 'Sex']

df_female = df[['Age Group', 'Number of diagnosis Female', 'Diagnosis Percentage Female']].copy()
df_female['Sex'] = 'Female'
df_female.columns = ['Age Group', 'Number of diagnosis', 'Ratio', 'Sex']

# Concatenate the two dataframes
df_table6_transposed = pd.concat([df_male, df_female])

# Display the final dataframe
df_table6_transposed
```

```python
# Load the file - Table 10
file_path_t10 = 'Tabla_10_diagnosis_age_method.xlsx'
xls_t10 = pd.ExcelFile(file_path_t10)

# Load the first sheet
df_t10 = pd.read_excel(xls_t10, sheet_name='Hoja1')
```

```python
# Remove columns that end with "2024"
df_filtered = df_t10.loc[:, ~df_t10.columns.str.endswith('2024')]

# Reshape the dataframe into the desired format: Age Group, Transmission Category, Number of diagnosis
df_transposed_t10 = df_filtered.melt(id_vars='Age Group', 
                                   var_name='Transmission Category', 
                                   value_name='Number of diagnosis')
```

```python
# Rename the transmission categories to match the given categories
transmission_mapping = {
    'Sexual Aggregated': 'Sexual',
    'Blood-borne Aggregated': 'Blood-Borne',
    'IDU Aggregated': 'IDU',
    'Vertical Aggregated': 'Vertical',
    'Not specified Aggregated': 'Not Specified'
}

df_transposed_t10['Transmission Category'] = df_transposed_t10['Transmission Category'].replace(transmission_mapping)
```

```python
# export tables
export_path = '/Tabular data/2024Q1/04 Transposed'

# file names
file_name_t6 = 'df_table6_age.csv'
file_name_t10 = 'df_table10_age.csv'

# Full path
full_export_path_t6 = f'{export_path}/{file_name_t6}'
full_export_path_t10 = f'{export_path}/{file_name_t10}'

df_table6_transposed.to_csv(full_export_path_t6, index=False)

df_transposed_t10.to_csv(full_export_path_t10, index=False)
```
