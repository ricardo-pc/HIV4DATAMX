### Demographic factor: Age

```python
# Import transformed raw files

df_table6 = pd.read_csv(f"{files_directory}/df_table6_age.csv")
df_table10 = pd.read_csv(f"{files_directory}/df_table10_age.csv")
```

Data explanation:
- df_table6: Contains age group, number of diagnoses and sex. Aggregated data from 1983 - 2024 Q1 in Mexico.
- df_table10: Contains data about the age group, tranmission category and the number of diagnoses. Aggregated data from 1983 - 2024 Q1 in Mexico.

```python
# Now let's do a basic exploratory analysis

import matplotlib.pyplot as plt

# Descriptive statistics
df_table10[df_table10['Age Group'] == '25 - 29'].describe()
```

```python
# Plotting the number of diagnoses by age group for df_table10
plt.figure(figsize=(10, 6))
plt.bar(df_table10['Age Group'], df_table10['Number of diagnosis'], color='blue')
plt.title('Number of Diagnoses by Age Group (Table 10)')
plt.xlabel('Age Group')
plt.ylabel('Number of Diagnoses')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the number of diagnoses by age group for df_table6 (split by Sex)
plt.figure(figsize=(10, 6))
for sex in df_table6['Sex'].unique():
    subset = df_table6[df_table6['Sex'] == sex]
    plt.bar(subset['Age Group'], subset['Number of diagnosis'], label=f"Sex: {sex}")

plt.title('Number of Diagnoses by Age Group and Sex (Table 6)')
plt.xlabel('Age Group')
plt.ylabel('Number of Diagnoses')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Create improved bar plot including Transmission Category
improved_bar_chart_with_category(
    df_table10, 'Age Group', 'Number of diagnosis', 'Transmission Category',
    title='Number of Diagnoses by Age Group and Transmission Category (Table 10)',
    xlabel='Age Group', ylabel='Number of Diagnoses'
)
```

Observations:
- Both datasets show that the number of diagnoses tends to increase with age, peaking around the teenage or early adult years. This suggests that younger adults and adolescents are a key demographic in HIV diagnoses.
- In the third graph, transmission categories, particularly “Sexual” transmission, dominates in all age groups, especially in the 15-19 and 25-29 age groups. This reflects a critical period of risk for HIV transmission due to behavioral factors.
- The “Vertical” transmission (from mother to child) primarily affects the youngest age groups, as expected. This category sees a decline as the age groups increase, indicating that vertical transmission is mostly a concern in infants and very young children.
- The second graph shows a clear male dominance in diagnoses, particularly in the older age groups (15-19, 20-24, etc.), where the male-to-female diagnosis ratio is much higher. This suggests that men, particularly young men, are at greater risk of HIV diagnoses.
